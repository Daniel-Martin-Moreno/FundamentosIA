```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
```

```python
# Cargar datos
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
print(X)
```

```python
# Estandarizar las características
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled)
```

# Sobre la estandarización
La estandarización transforma los datos para que cada variable tenga media 0 y desviación estándar 1. Esto es clave en PCA porque:

PCA se basa en varianzas y covarianzas.

Si las variables están en escalas distintas (por ejemplo, centímetros y gramos), unas dominarían sobre otras solo por su magnitud.

```python
# Aplicar PCA (sin limitar componentes)
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Varianza explicada por cada componente
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

# Visualizar curva de varianza explicada
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='o', linestyle='--', color='b')
plt.title('Curva de Varianza Explicada Acumulada por PCA')
plt.xlabel('Número de Componentes Principales')
plt.ylabel('Varianza Explicada Acumulada')
plt.grid(True)
plt.xticks(range(1, len(cumulative_variance)+1))
plt.show()
```

```python
# Crear un DataFrame legible con las cargas de cada componente
loadings = pd.DataFrame(pca.components_.T,  # transpuesta para tener variables como filas
                        columns=[f'PC{i+1}' for i in range(len(pca.components_))],
                        index=feature_names)

print(loadings)
```
